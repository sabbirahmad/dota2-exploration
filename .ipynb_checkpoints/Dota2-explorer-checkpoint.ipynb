What if I do an weighted average of densities by action types, then we can combine peaks. In the process, we produce a distribution by signifcance of events. Then we run mean-shift on the final distribution.

```python
x = sample[0]
y = sample[1]

xmin = x.min()
xmax = x.max()
ymin = y.min()
ymax = y.max()

X, Y = np.mgrid[xmin:xmax:100j, ymin:ymax:100j]
positions = np.vstack([X.ravel(), Y.ravel()])
    
def produce_kde_by_action(action):
    print(action)
    sample = original[original['action'] == action][['posX', 'posY']].as_matrix().transpose()
    kde = gaussian_kde(sample)

    Z = np.reshape(kde(positions).T, X.shape)
    
    return (kde, Z)

res = [produce_kde_by_action(action) for action in original['action'].unique()]
```

```python
kdes, Zs = zip(*res)
```

```python
Z_ave = np.mean(Zs, axis=0)

plt.imshow(np.rot90(Z_ave), cmap=plt.cm.viridis,
          extent=[xmin, xmax, ymin, ymax])
```

```python
Z_ave = np.percentile(Zs, 100, axis=0)

plt.imshow(np.rot90(Z_ave), cmap=plt.cm.viridis,
          extent=[xmin, xmax, ymin, ymax])
```

```python
prob_max = np.min([np.max(z) for z in Zs])
Zs_ = [z*prob_max/np.max(z) for z in Zs]

Z_ave = np.percentile(Zs_, 100, axis=0)
Z_ave = Z_ave/np.sum(Z_ave)

plt.imshow(np.rot90(Z_ave), cmap=plt.cm.viridis,
          extent=[xmin, xmax, ymin, ymax])
```

```python
cut_of = Z_ave.min() + 0*(Z_ave.max() - Z_ave.min())/10
```

```python
plt.scatter(X[Z_ave > cut_of], Y[Z_ave > cut_of], s=0.1)
```

```python
T = Z_ave > cut_of
p = np.asarray(Z_ave[T])**10
p = p/np.sum(p)

points = list(zip(X[T], Y[T]))
s_points = [points[
        np.random.choice(len(p), p=p)
    ] for i in range(0, 50000)]
```

```python
x_, y_ = zip(*s_points)
```

```python
plt.scatter(x_, y_, s=10, alpha=0.01)
```

```python
ms = MeanShift(bandwidth=1000, bin_seeding=True)
ms.fit(s_points)
```

```python
labels = ms.labels_
cluster_centers = ms.cluster_centers_
```

```python
x, y = zip(*cluster_centers)
plt.scatter(x, y)
```

```python
original['cluster_Z_ave'] = ms.predict(
    [list(x) for x in original[['posX', 'posY']].values]
)
```

```python
sns.lmplot("posX", "posY", data=original, x_jitter=100, y_jitter=100, hue="cluster_Z_ave", fit_reg=False, scatter_kws={"s": 5, "alpha": 0.025}, size=8)
```

### Summarize the process:

1. estimate KDE for each action
2. compute probabilities for all data points over all KDEs
3. obtain the average KDE: 
        prob_max = the min of (max per KDE) over all KDEs
        Zs_ = adjust each KDE to have max prob as prob_max
        Z_ave = assign the max probability over all KDEs point-wise
        normalize Z_ave

4. Find the cut of from the KDE to filter points (70% percentile); filter the data
5. Run MinShift on the filtered dataset: obtain the centroids
6. Map the centroid labels to each data point from the filtered set
7. All other points are neutral
